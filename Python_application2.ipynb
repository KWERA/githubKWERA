```python
#1.1. Store the data for each district in a NumPy array.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics as stats

Kampala = [1200, 1250, 1300, 1350, 1420, 1500, 1580, 1650, 1720, 1800]
Wakiso = [950, 1000, 1070, 1150, 1220, 1300, 1390, 1480, 1570, 1670]
Gulu = [320, 330, 345, 360, 375, 390, 410, 430, 455, 480]

Kamp_array =np.array(Kampala)
Wak_array = np.array(Wakiso)
Gulu_array = np.array(Gulu)
print (Kamp_array)
print (Wak_array)
print (Gulu_array)
```

```python
#import statistics as stats
#print(dir(statistics))
```

```python

#1.2. Use the statistics module to compute mean, median, variance, and standard deviation for each district.

def stats_function(data):
    District_mean = round(stats.mean(data),3)
    District_median = round(stats.median(data),3)
    District_variance = round(stats.variance(data),3)
    District_SD = round(stats.stdev(data),3)
    return District_mean, District_median, District_variance,District_SD

print("Kampala Statistics(mean, median, variance,Standard Deviation): \n", stats_function(Kamp_array))
print( '----'*10)
print("Wakiso Statistics(mean, median, variance,Standard Deviation): \n", stats_function(Wak_array))
print( '----'*10)
print("Gulu Statistics(mean, median, variance,Standard Deviation): \n", stats_function(Gulu_array))
print( '----'*10)
```

```python
#1.3. Generate a Fibonacci sequence of length 5 to project the next 5 years of growth.
print("\n 1.3. Generate a Fibonacci sequence of length 5 to project the next 5 years of growth. \n")

def fibonacci_predict(arr, n_future=5, scale=1.0):
    pred = arr.copy().tolist()
    for _ in range(n_future):
        next_val = (pred[-1] + pred[-2]) * scale
        pred.append(next_val)
    return np.array(pred)

Kamp_pred = fibonacci_predict(Kamp_array,5)
wak_pred = fibonacci_predict(Wak_array,5)
Gulu_pred = fibonacci_predict(Gulu_array,5)
                              
print("Predicted population for Kampala:", Kamp_pred[-5:])
print("Predicted population for Wakiso:", wak_pred[-5:])
print("Predicted population for Gulu:", Gulu_pred[-5:])
```

```python
Kamp_array2 = Kamp_pred[-5:]
Wak_array2 = wak_pred[-5:]
Gulu_array2 = Gulu_pred[-5:]

def stats_function2(data, projected_data):
    district_var = round(stats.variance(data),2)
    district_proj_var2 = round(stats.variance(projected_data),2)
    return district_var,district_proj_var2

print("Kampala Statistics(variance compared with Projected variance): \n", stats_function2(Kamp_array,Kamp_array2))
print( '----'*10)
print("Wakiso Statistics(variance compared with projected variance): \n", stats_function2(Wak_array,Wak_array2))
print( '----'*10)
print("Gulu Statistics(variance compared with projected variance): \n", stats_function2(Gulu_array,Gulu_array2))
print( '----'*10)
```

```python
'''Question 3: Lake Victoria Fish Export Risk Model
Attempt the following tasks:
3.1. Generate 15 Fibonacci numbers to simulate fish stock growth.
3.2. Simulate fish prices in UGX per kg, e.g. [12000, 12500, 11800, 13000, 12800,...].
3.3. Multiply stock × price to estimate daily revenue.
3.4. Compute mean, median, variance of revenue using statistics.
3.5. Print 'High risk' if variance > 50,000, else 'Low risk'.
3.6. Plot fish stock vs revenue trend.'''

# 3.1. Generate 15 Fibonacci numbers to simulate fish stock growth.
print(' 3.1. Generate 15 Fibonacci numbers to simulate fish stock growth. ')
print("-"*80,end = '\n')

# Generate 15 Fibonacci numbers
def fibonacci(n):
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[-1] + fib[-2])
    return fib

# Generate and print
fish_stock = fibonacci(15)
print("Simulated Fish Stock Growth (Fibonacci sequence): \n")
print(fish_stock, '\n')

# 3.2. Simulate fish prices in UGX per kg, e.g. [12000, 12500,...].14k-17k
print("-"*80)
print('3.2. Simulate fish prices in UGX per kg','\n')

KGprice = np.random.randint(14000, 17000, size=15)
print('Simulate fish prices in UGX :\n', KGprice,'\n')
#print("-"*40)

#3.3. Multiply stock × price to estimate daily revenue.
print('3.3. Multiply stock × price to estimate daily revenue. ')
print("-"*80, end='\n')

DailyRevenue = fish_stock * KGprice
print ('Calculated daily revenue: \n', DailyRevenue, end='\n\n' )

#3.4. Compute mean, median, variance of revenue using statistics.
print( "3.4. Compute mean, median, variance of revenue using statistics. ")
print("-"*80,'\n')
mean_revenue = np.mean(DailyRevenue)
print('Mean revenue: ',mean_revenue,'\n')

Median_revenue = np.median(DailyRevenue)
print('Median revenue: ',Median_revenue,'\n')

variance_Rev= np.var(DailyRevenue)
print('Revenue variance: ',variance_Rev,'\n')

#3.5. Print 'High risk' if variance > 50,000, else 'Low risk'.
print("3.5. Print 'High risk' if variance > 50,000, else 'Low risk'")
print("-"*80,'\n')
if variance_Rev>50000:
    print( 'High risk')
else:
    print('Low risk')
```
